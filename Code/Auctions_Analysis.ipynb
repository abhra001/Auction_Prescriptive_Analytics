```python
mu=np.mean(swoopo["retail"])

swoopo['reward_75']=((swoopo["retail"]-mu)**2)/10000

swoopo['reward_60']=2*np.mean(swoopo['reward_75'])-((swoopo["retail"]-mu)**2)/10000

reward=[]
for i in swoopo.index:
    if swoopo["bidfee"][i]==75:
        reward=reward+[swoopo["reward_75"][i]]
    else:
        reward=reward+[swoopo["reward_60"][i]]
    
swoopo["Reward"]=reward;
```

```python
from sklearn.model_selection import train_test_split

swoopo['treatment'] = np.where(swoopo['bidfee']==75, 1, 0)
swoopo['revenue'] = swoopo['bidfee']*swoopo['placedbids']

swoopo_train, swoopo_test = train_test_split(swoopo, test_size=0.4, random_state=1)
```

```python
# Model treatment effect

feature_names = ["bidincrement","retail","flg_click_only","flg_beginnerauction","flg_fixedprice","flg_endprice"]

# standard_mse is an MSE criterion where nodes store treatment effects
tree1 = CausalTreeRegressor(criterion='standard_mse', 
                            control_name=0,
                            min_impurity_decrease=0,
                            min_samples_leaf=200, 
                            leaves_groups_cnt=True)
tree1.fit(X=swoopo_train[feature_names].values,
          treatment=swoopo_train['treatment'].values,
          y=swoopo_train['revenue'].values
          )


tree2 = CausalTreeRegressor(criterion='causal_mse',
                            control_name=0,
                            min_samples_leaf=200, 
                            leaves_groups_cnt=True)
tree2.fit(X=swoopo_train[feature_names].values,
          treatment=swoopo_train['treatment'].values,
          y=swoopo_train['revenue'].values
          )
```

```python
tree1_ite_pred = tree1.predict(swoopo_test[feature_names].values)
tree2_ite_pred = tree2.predict(swoopo_test[feature_names].values)

swoopo_result = pd.DataFrame(
    {
        'tree_mse_ite': tree2_ite_pred,
        'tree_causal_mse_ite': tree1_ite_pred,
        'outcome': swoopo_test['revenue'],
        'is_treated': swoopo_test['treatment'],
    }
)
swoopo_result = swoopo_result.reset_index(drop=True)
swoopo_result
```
